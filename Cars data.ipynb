Q) Find the car with Best Power to Weight Ratio
 Create a new a column "Power_to_weight" =Horsepower/weight ,
 and find the car with highest ratio

```python
data['Power_to_Weight']=data['Horsepower']/data['Weight']
```

```python
highest_power_to_weight_car = data.loc[data['Power_to_Weight'].idxmax(), ['Make','Model', 'Power_to_Weight']]
# The error was caused by referencing 'highest_power_to_weight_car' before it was assigned a value.
# Changed 'highest_power_to_weight_car' to 'Power_to_Weight' in the column selection to access the desired column.
```
